```python
#SVM활용하여 학업성과예측 모델 구현
# 2018. 6.18 박희순
# 데이터 셋 : 2013년 ~ 2017년 학습데이터 
#gradua,schoarYn,Access,Attend,Activity,LecCnt,StepTime,Homework,mobileRate,PassRate,GrateAvg
#최종학력,경제적 부담,로그인횟수,학습시간,학습활동,총강의갯수,총강의시간,학습활동 개수,모바일사용환경,이수율,평균평점

from sklearn import svm, metrics
import random, re


#학습분석 정보 데이터 읽어 들이기
csv = []

#타이틀 필요
with open('./dataset/studyset_2013_2017_fact11.csv', 'r', encoding='utf-8') as fp:
    for line in fp:
        line = line.strip() #양쪽 공백 제거
        cols = line.split(',')
        fn = lambda n : float(n) if re.match(r'^[0-9\.]+$',n) else n
        cols = list(map(fn, cols))
        csv.append(cols)
del csv[0] #헤더정보 삭제하기


#데이터 섞기
random.shuffle(csv)
#csv  
#학습데이터와 테스트전용 데이터로 분할 하기 2:1

total_len = len(csv)
train_len= int(total_len * 2 / 3)
train_data = []
train_label = []
test_data = []
test_label = []

#print(train_len)
#print(total_len)



for i in range(total_len):
    data = csv[i][0:11]
    label = csv[i][11:12]

    #print( i )
    
    if i < train_len:
        train_data.append(data)
        train_label.append(label)
        #print( "train")
        #print(data)
        #print(train_label[i])
    else:
        test_data.append(data)
        test_label.append(label)
        

#print(test_data)

#임의 데이터 입력

clf = svm.SVC()
clf.fit(train_data, train_label)
pre = clf.predict(test_data)  

for i in range(total_len-train_len):
    print(test_data[i], test_label[i], pre[i])
    
    if test_label[i][0] == pre[i]:
        print(i,test_data[i], test_label[i][0],'정답', pre[i])
    else:
        print(i,test_data[i], test_label[i][0],'오답', pre[i])

ac_score = metrics.accuracy_score(test_label, pre)

print("정답율" , ac_score)
```
